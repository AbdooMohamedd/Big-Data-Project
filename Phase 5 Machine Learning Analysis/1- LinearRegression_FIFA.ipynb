```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = r"C:\Users\dell\Desktop\Full cleand fifa players 2.csv"
df = pd.read_csv(file_path)
```

```python
# Selecting the target variable and features
target_variable = 'Pace'
features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
features.remove(target_variable)  # Remove the target variable from the feature list
```

```python
# Splitting the data into training and testing sets
X = df[features]
y = df[target_variable]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Standardizing the feature data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Training a Linear Regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
# Predicting and evaluating the model
y_pred = model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)
```

```python
# Printing the evaluation metrics
print("Mean Squared Error (MSE):", mse)
```

```python
# Plotting the actual vs predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values for Pace')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red')  # Diagonal line
plt.show()
```

# Data Cleaning

```python
df = pd.read_csv(r"Full cleand fifa players.csv", sep=";")
df.head()
```

```python
# check for shape of the data
shape = list(df.shape)
print(shape)
print(f"the dataframe has {shape[0]} Rows and {shape[1]} columns.")
```

```python
# Step 1: Remove duplicate rows
dff=df
dff = dff.drop_duplicates()
```

```python
# Step 2: Remove rows where 'Club' column is null
dff = dff.dropna(subset=['Club'])
```

```python
# # Step 3: Remove unnecessary columns
# columns_to_drop = ['Player Card Image']
# dff = dff.drop(columns=columns_to_drop)
```

```python
# Step 4: Replace null values in the 'gender' column with 'Male'
dff['gender'] = dff['gender'].fillna('Male')
```

```python
# step 5:Normalizing text data for 'Name', 'Club', 'Nation', and 'League' columns
dff['Name'] = dff['Name'].str.title()
dff['Club'] = dff['Club'].str.title()
dff['Nation'] = dff['Nation'].str.title()
dff['League'] = dff['League'].str.title()

# Show the head of the DataFrame to confirm text normalization
dff[['Name', 'Club', 'Nation', 'League']].head()
```

```python
# step 6:Cleaning 'Skill Moves' and 'Weak Foot' columns by removing the star and trailing whitespace

# Then converting the columns to integer type
dff['Skill Moves'] = dff['Skill Moves'].str.replace(' \\u2605', '').astype(int)
dff['Weak Foot'] = dff['Weak Foot'].str.replace(' \\u2605', '').astype(int)

# Display the head of the dataframe to confirm the changes
dff[['Skill Moves', 'Weak Foot']].head()
```

```python
# step 7: delete all random characters and diffrent languages  from these columns
columns_to_clean = ['Name', 'Card Title', 'Club', 'Nation', 'League', 'Foot', 'AcceleRATE', 'Height', 'Att/Def. WR', 'AcceleRATE_1']

# Adjusting the regex pattern to keep English letters, numbers, whitespace, and the characters \, /, |
regex_pattern = r'[^a-zA-Z0-9 \\\/\|\"\']'

# Applying the adjusted regex pattern to the specified columns
for col in columns_to_clean:
    dff[col] = dff[col].astype(str).str.replace(regex_pattern, '', regex=True)

# Display the head of the dataframe to confirm the changes
dff[columns_to_clean].head()
```

# leveraged to check for outliers, patterns, and trends in some of the given data

```python
cat_cols=dff.select_dtypes(include=['object']).columns
num_cols = dff.select_dtypes(include=np.number).columns.tolist()
for col in num_cols:
    print(col)
    print('Skew :', round(dff[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    dff[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=dff[col])
    plt.show()
```

```python
# Identify numerical columns with continuous data
numerical_columns = ['Pace', 'Acceleration', 'Sprint Speed', 'Shooting', 'Passing', 'Dribbling', 'Defending', 'Physicality', 'Penalties']

# Use z-scores to detect outliers
z_scores = pd.DataFrame()
for column in numerical_columns:
    z_scores[column] = (dff[column] - dff[column].mean()) / dff[column].std()

# Set a z-score threshold for outlier detection
z_threshold = 3
outliers = (z_scores > z_threshold) | (z_scores < -z_threshold)

# Create box plots for relevant attributes
plt.figure(figsize=(12, 6))
sns.boxplot(data=dff[numerical_columns], palette="Set2")
plt.title('Box Plots for Numerical Attributes')
plt.show()
```

```python
# Plot histograms or kernel density plots
plt.figure(figsize=(12, 8))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(dff[column], kde=True, color='skyblue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
```

```python
# Selecting 36 numeric columns from the DataFrame
numeric_columns = dff.select_dtypes(include='number').columns[:36]

# Creating a subplot grid (6 rows x 6 columns)
plt.figure(figsize=(18, 18))
plt.suptitle('Distribution of Numeric Columns', fontsize=16)

for i, column in enumerate(numeric_columns, 1):
    plt.subplot(6, 6, i)
    sns.histplot(dff[column], kde=True, color='skyblue')
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# correlations for fisrt 30 columns
subset_columns = dff.iloc[:, :30]

correlation_matrix = subset_columns.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix (First 10 Columns)')
plt.show()
```

```python
# Selecting 36 numeric columns from the DataFrame
numeric_columns = dff.select_dtypes(include='number').columns[:36]

# Creating a subplot grid (6 rows x 6 columns)
fig, axes = plt.subplots(nrows=6, ncols=6, figsize=(18, 18))
fig.suptitle('Scatter Plots for Numeric Columns', fontsize=16)

# Iterating over each numeric column and creating scatter plots
for i, col in enumerate(numeric_columns):
    row_index = i // 6
    col_index = i % 6
    axes[row_index, col_index].scatter(dff[col], dff['Pace'], alpha=0.5)
    axes[row_index, col_index].set_title(col)
    axes[row_index, col_index].set_xlabel(col)
    axes[row_index, col_index].set_ylabel('Pace')

# Adjusting layout and displaying the plot
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

## in this phase we will insert the CSV file that has FIFA 24 players data into MongoDB using Python

Mongo Before the insertion


# import the libraries

```python
import pandas as pd
from pymongo import MongoClient
```

# Read the data 

```python
# Load the dataset into a pandas DataFrame
file_path = r"Full cleand fifa players 2.csv"
data = pd.read_csv(file_path, sep=';')
data
```

# connect to MongoDB and insert the Data

```python
client = MongoClient("mongodb://localhost:27017/")
client.list_database_names()
```

```python
# Connect to MongoDB
client = MongoClient('localhost', 27017) 
db = client['fifa_database']
collection = db['fifa_players']

# Convert the DataFrame to a list of dictionaries
data_dict = data.to_dict('records')

# Upload the data to MongoDB
collection.insert_many(data_dict)
```

MongoDB After The insertion


# Read the inserted Data from MongoDB

```python
# List all collection names in the database
db = client['fifa_database']
collection_names = db.list_collection_names()
collection_names
```
